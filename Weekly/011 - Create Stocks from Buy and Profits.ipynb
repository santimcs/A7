```python
# Adding specified columns to the DataFrame
columns_data_types = {
    'max_price': 'float64',
    'min_price': 'float64',
    'buy_target': 'int64',
    'sell_target': 'float64',
    'volume': 'float64',
    'beta': 'float64',
    'qty': 'int64',
    'buy_spread': 'int64',
    'sell_spread': 'int64',
    'bl': 'int64',
    'sh': 'int64',
    'reason': 'object',
    'market': 'object'
}

for col, dtype in columns_data_types.items():
    df_uni[col] = '0'
    df_uni[col] = df_uni[col].astype(dtype)

df_uni.head()  # Displaying the first few rows of the updated DataFrame
```

```python
df_uni['buy_spread'] = -4
df_uni['sell_spread'] = 4
df_uni['reason'] = 'XXX'
df_uni
```

```python
names = df_uni['name'].values.tolist()
in_p = ", ".join(map(lambda name: "'%s'" % name, names))
in_p
```

```python
sql = """
SELECT name, market 
FROM tickers
WHERE name in (%s)"""
sql = sql % in_p
tickers = pd.read_sql(sql, conmy)
tickers
```

```python
# Directly merging on 'Name' column, since it's equivalent to 'name' column in df_merge
df_merge = df_uni.merge(tickers, on='name', how='left')
df_merge.dtypes
```

```python
# Dropping the market_x column and renaming market_y to market
df_merge.drop(columns=['market_x'], inplace=True)
df_merge.rename(columns={'market_y': 'market'}, inplace=True)
df_merge[['name', 'market']]
```
