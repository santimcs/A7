```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

data_path = "../data/"
csv_path = "\\Users\\User\\iCloudDrive\\"
one_path = "\\Users\\User\\OneDrive\\Documents\\Data\\"
osd_path = "\\Users\\User\\OneDrive\\Documents\\obsidian-git-sync\\Data\\"
```

```python
url = 'https://www.bbc.com/sport/football/tables'
r = requests.get(url)
html = r.text
#html
```

```python
soup = BeautifulSoup(html, 'html.parser')
#soup
```

```python
table = soup.find('table')
df = pd.read_html(str(table))[0]
df
```

```python
#print(df.head())
```

```python
df.columns
```

```python
df.drop(columns=['Unnamed: 0','Unnamed: 1','Form'],inplace=True)
df
```

```python
df.drop([20],inplace=True)
df
```

```python
file_name = 'Premier-League.csv'
data_file = data_path + file_name
output_file = csv_path + file_name
osd_file = osd_path + file_name
one_file = one_path + file_name

df.sort_values(['Pts'],ascending=[False]).to_csv(output_file, index=False)
df.sort_values(['Pts'],ascending=[False]).to_csv(data_file, index=False)
df.sort_values(['Pts'],ascending=[False]).to_csv(osd_file, index=False)
df.sort_values(['Pts'],ascending=[False]).to_csv(one_file, index=False)
```
